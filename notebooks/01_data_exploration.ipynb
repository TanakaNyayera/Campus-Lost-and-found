```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Matthew .K. Maunga\OneDrive\Desktop\MSU project\MSU_Lost_Found_ML_System\data\raw\descriptions.csv")

print(f"Total records: {len(df)}")
print(f"Columns: {list(df.columns)}")
print(f"\nFirst 5 records:")
df.head()
```

```python
print(f"Total records: {len(df)}")
print(f"\nStatus distribution:")
print(df['status'].value_counts())
print(f"\nCategory distribution:")
print(df['category'].value_counts())
print(f"\nAny missing values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Category distribution chart
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1 - Category distribution
df['category'].value_counts().plot(
    kind='bar', ax=axes[0], color='steelblue', edgecolor='black'
)
axes[0].set_title('Number of Items per Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Chart 2 - Lost vs Found
df['status'].value_counts().plot(
    kind='bar', ax=axes[1], color=['tomato', 'seagreen'], edgecolor='black'
)
axes[1].set_title('Lost vs Found Distribution')
axes[1].set_xlabel('Status')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../reports/figures/data_distribution.png')
plt.show()
print(" Chart saved to reports/figures/")
```
